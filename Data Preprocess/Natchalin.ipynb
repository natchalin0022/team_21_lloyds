```python
import requests
from requests.auth import HTTPBasicAuth

API_KEY = "37037e79-651d-4621-95f5-85db5a2842ea"
BASE_URL = "https://api.company-information.service.gov.uk"

def get(endpoint, params = None):
    response = requests.get(
        f'{BASE_URL}{endpoint}',
        auth = HTTPBasicAuth(API_KEY, ''),
        params = params
    )
    response.raise_for_status()
    return response.json()
```

## Search for Companies - from name

```python
def search_companies(query, items_per_page = 20, start_index = 0):
    return get('/search/companies', {'q':query, 'items_per_page': items_per_page, 'start_index': start_index})
```

```python
results = search_companies('Lloyds Banking Group', items_per_page = 5)
for r in results['items']:
    print(r['title'], r['company_number'])
```

## SIC Codes related

```python
def search_companies_advanced(**kwargs):
    return get('/advanced-search/companies', kwargs)
```

```python
def get_companies_in_industry(industry_code):
    filter_sic = search_companies_advanced(
        sic_codes = f'{industry_code}',
        company_type = 'ltd', # SME
        company_status = 'active',
        size = 100
    )
    for d in filter_sic['items']:
        print(d['company_name'])
```

```python
get_companies_in_industry(int(get_company('SC095000')['sic_codes'][0]))
```
