```python
import json
import os
from pathlib import Path
```

```python
json_path = Path.cwd() / "ocr_results.json"
json_path
```

```python
with open(json_path, "r") as f:
    data = json.load(f)


for index, item in data.items():
    data[index] = json.loads(item)
```

```python
data
```

```python
found_tables = {}

for i,x in data.items():
    if x['table_found'] == True: found_tables[i] = x
```

```python
found_tables
```

```python
table_keys = list(found_tables.keys())
```

```python
ree = 7

print(table_keys[ree])

for i,x in enumerate(found_tables[table_keys[ree]]["tables"]):
    print(f"Table: {x['table_name']}")
    print(f"Header: {x['headers']}")
    print(f"Rows:")
    print("---------------------------------------------")
    for j,y in enumerate(x['rows']):
        print(f"Row {j+1}: {y}")
        
```
